```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.animation import FuncAnimation

# Graph Attributes
x = np.arange(5)
y = [20, 25, 30, 35, 40]  # Example city-mpg values
z = [0]*5  # Set all bars to start from z=0

colors = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99', '#c299ff']  # Bar colors

# Bar Attributes
dx = [0.4]*5  # Width of the bar
dy = [2]*5    # Length of the bar
dz = [15000, 18000, 20000, 22000, 25000]  # Prices for each body style

# Set up the plot
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

# Initial Bar Plot
bars = ax.bar3d(x, y, z, dx, dy, dz, color=colors, zsort='average')

# Set labels
ax.set_xlabel('Body Style', labelpad=12, color='white')
ax.set_ylabel('City MPG', color='white')
ax.set_zlabel('Price', labelpad=12, color='white')

# Set background and grid color
ax.set_facecolor((0.6, 0.6, 0.6, 1))
plt.gcf().patch.set_facecolor((0.6, 0.6, 0.6, 1))

ax.set_title("Change of price with body style and city-mpg", color='white')
ax.tick_params(axis='x', labelcolor='white')
ax.tick_params(axis='y', labelcolor='white')
ax.tick_params(axis='z', labelcolor='white')

ax.set_xticks(x)
ax.set_xticklabels(['Convertible', 'Hatchback', 'Sedan', 'Wagon', 'Hardtop'])

ax.view_init(elev=20, azim=45)  # Adjust the viewing angle for better perspective

animate_rotation = True  # Set this to True for rotation, False for static 3D plot

if animate_rotation:
    def update(frame):
        ax.view_init(elev=20, azim=frame)  
        
    ani = FuncAnimation(fig, update, frames=np.arange(0, 360, 1), interval=50)
else:
    pass

plt.show()
```
