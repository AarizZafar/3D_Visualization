```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('DataSet/auto_clean.csv')
df.head(3)
```

```python
%matplotlib widget
```

```python
plt.figure(figsize=(8,8))
ax = plt.axes(projection='3d')
fg = ax.scatter3D(df['length'], df['width'], df['height'],
                 s = df['price']/60, c = df['peak-rpm'])

ax.set_xlabel('length', color='white')
ax.set_ylabel('widht', color='white')
ax.set_ylabel('Height', color='white')

ax.tick_params(axis='x', labelcolor='white')
ax.tick_params(axis='y', labelcolor='white')
ax.tick_params(axis='z', labelcolor='white')

# ax.set_facecolor((0,0,0,0.5))
# plt.gcf().patch.set_facecolor((0,0,0,0.5))

ax.set_facecolor((0.7, 0.7, 0.7, 1))  
plt.gcf().patch.set_facecolor((0.7, 0.7, 0.7, 1))

plt.colorbar(fg)

plt.show()
```
